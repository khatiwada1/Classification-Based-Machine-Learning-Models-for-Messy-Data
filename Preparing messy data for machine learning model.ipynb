```python
# Before any thing else lets load some basic python libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import warnings
warnings.filterwarnings("ignore")
```

```python
#now read the training data 
tst_raw=pd.read_csv('exercise_40_test.csv')
tst_raw.head(10)
#print(tst_raw.shape)
```

```python
#lets define Categorical data
tst_raw_category=tst_raw.loc[:,tst_raw.dtypes==np.object]
#tst_raw_category.shape
tst_raw_category.head()
```

```python
#summarize categorical variables
tst_raw_category.describe(include='all')
```

```python
#based on this observation Column x39 and x99 are qunique with only one kind of values. So lets drop them
nunique = tst_raw_category.nunique()
cols_to_drop = nunique[nunique == 1].index
tst_raw_category = tst_raw_category.drop(cols_to_drop, axis=1)
tst_raw_category.describe(include='all')
```

```python
#lets find the null values in these categorical data and impute them with mode(max frequency)methos
for col in tst_raw_category:
    max_freq = tst_raw_category[col].value_counts().index[0]
    tst_raw_category[col][pd.isna(tst_raw_category[col])] = max_freq
```

```python
tst_raw_category.describe(include='all')
```

```python
#now lets remove % sign from column x7 and x19
tst_raw_category['x7'] = tst_raw_category['x7'].replace('\%','',regex=True).astype(float)
tst_raw_category['x19'] = tst_raw_category['x19'].replace('\$','',regex=True).astype(float)
tst_raw_category.head()
```

```python
#features X31 and X93 are binary class with Yes and no nominal values.Lets convert them into numerical values of 0 and 1
tst_raw_category['x31']=tst_raw_category['x31'].replace(['no','yes'],[0,1])
tst_raw_category['x93']=tst_raw_category['x93'].replace(['no','yes'],[0,1])
tst_raw_category.head()
```

```python
# Now lets work with days and lets make them uniform Ordinal 
tst_raw_category["x3"].replace({"Sat": "Saturday", "Sun": "Sunday", "Mon": "Monday", "Tue": "Tuesday",
                  "Wed": "Wednesday", "Thur": "Thursday", "Fri": "Friday"}, inplace=True)
```

```python
#Another nomial category is gender
tst_raw_category['x24']=tst_raw_category['x24'].replace(['female','male'],[0,1])
tst_raw_category.head()
```

```python
#Now I will convert them with dictionary frequency,save to a mapping function and map it back to dictionary 
#Tried label encoder it did not worked properly. 
freq=tst_raw_category.x3.value_counts().to_dict()
tst_raw_category.x3=tst_raw_category.x3.map(freq)
```

```python
# For States X33
freq1=tst_raw_category.x33.value_counts().to_dict()
tst_raw_category.x33=tst_raw_category.x33.map(freq1)
```

```python
##for months x60
freq2=tst_raw_category.x60.value_counts().to_dict()
tst_raw_category.x60=tst_raw_category.x60.map(freq2)
```

```python
#for months x65
freq3=tst_raw_category.x65.value_counts().to_dict()
tst_raw_category.x65=tst_raw_category.x65.map(freq3)
```

```python
#for months x77
freq4=tst_raw_category.x77.value_counts().to_dict()
tst_raw_category.x77=tst_raw_category.x77.map(freq4)
```

```python
tst_raw_category.head()
```

```python
#lets separate the data with numeric values only
tst_raw_num = tst_raw.loc[:,tst_raw.dtypes == np.float64]
tst_raw_num.head()
```

```python
# Now lets create a new dataset with all numerical values
tst_clean=pd.concat([tst_raw_num,tst_raw_category], axis=1)
```

```python
tst_clean.shape
```

```python
#summarize categorical variables
tst_clean.describe(include='all')
```

```python
# lets find the percentage of null values in each of the column from the clearner version of data
total =tst_clean .isnull().sum().sort_values(ascending=False)
miss_perc = tst_clean.isnull().sum()/tst_clean.isnull().count()*100
percent= (round(miss_perc, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', '%'])
missing_data
```

Without bclearly kowing the physical meaning of all these columns and trying to impute allthese columns with nemerical values is not making sense. So, an choosing an arbitarary percentage cut off of 20% and will drop the columns that has missing values of more than 20%.

```python
#dropping columns with missing values more than 20%
tst_df = tst_clean.loc[:, tst_clean.isnull().mean() < .20]
```

```python
tst_df.shape
```

```python
#Now lets do the numerical imputation for all numerical values using iterative imputer. Doing multiple imputation 
#does not make sense here since we dont know what these varaible and features really mean>

from fancyimpute import IterativeImputer

#calling MICE imputer
mice_imp = IterativeImputer()
#now impute missing values with MICEimputer
imp_tst_df = pd.DataFrame(mice_imp.fit_transform(tst_df))
imp_tst_df.index = tst_df.index
imp_tst_df.columns = tst_df.columns
```

```python
# lets check with simple person correlation and see if any columns are highly correlated

def correlation(df, threshold):
    col_corr=set()
    corr_matrix=df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
corr_features=correlation(imp_tst_df, 0.6)
#len(set(corr_features))
corr_features
```
